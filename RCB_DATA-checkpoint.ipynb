```python
plt.figure(figsize=(10,6))
tenure = rcb_captain['To'] - rcb_captain['From']
plt.bar(rcb_captain['Player'],tenure)
plt.xlabel('Captain')
plt.ylabel('Tenure(Years)')
plt.title('RCB Captain Tenure')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

captains = rcb_captain['Player']
wins = rcb_captain['Won']
losses = rcb_captain['Lost']
win_percentages = rcb_captain['Win%']


x = range(len(captains))
width = 0.35

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(x, wins, width, label='Wins', color='green')
ax1.bar([i + width for i in x], losses, width, label='Losses', color='red')
ax1.set_xlabel('Captain')
ax1.set_ylabel('Number of Matches', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xticks([i + width / 2 for i in x])
ax1.set_xticklabels(captains, rotation=45, ha='right')
ax1.legend(loc='upper left')


ax2 = ax1.twinx() 

ax2.plot(x, win_percentages, color='blue', marker='o', label='Win Percentage')
ax2.set_ylabel('Win Percentage', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.legend(loc='upper right')

plt.title('RCB Captain Wins, Losses, and Win Percentage')
plt.tight_layout()
plt.show()
```
